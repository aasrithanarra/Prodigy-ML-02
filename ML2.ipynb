```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset from CSV (assuming you uploaded customer_data.csv to Colab session)
df = pd.read_csv('customer_data.csv')

# Set CustomerID as the index
df.set_index('CustomerID', inplace=True)

# Number of clusters
k = 3

# Create KMeans model
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model
kmeans.fit(df)

# Get cluster labels
df['Cluster'] = kmeans.labels_

print(df)

# Plotting the clusters
plt.scatter(df['Grocery'], df['Clothing'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Grocery Spending')
plt.ylabel('Clothing Spending')
plt.title('Customer Clusters')
plt.show()
```
